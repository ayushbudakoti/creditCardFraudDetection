```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dataset = pd.read_csv('creditcard.csv')
```

## Data exploration

```python
dataset.head()
```

```python
dataset.isna()
```

## Data Pre-processing

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
dataset['normalizedAmount'] = sc.fit_transform(dataset['Amount'].values.reshape(-1,1))
dataset = dataset.drop(['Amount'], axis = 1)
```

```python
dataset.head()
```

```python
dataset = dataset.drop(['Time'], axis = 1)
dataset.head()
```

```python
X = dataset.iloc[:, dataset.columns != 'Class'].values
y = dataset.iloc[:, dataset.columns == 'Class'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.3, random_state = 0)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Model Preparation

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
```

```python
random_forest.fit(X_train, y_train)
```

```python
y_pred = random_forest.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(cm)
```

```python
sns.countplot(x = 'Class', data=dataset)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X_resample, y_resample = SMOTE().fit_sample(X,y)
```

```python
X_resample = pd.DataFrame(X_resample)
y_resample = pd.DataFrame(y_resample)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resample, y_resample, test_size = 0.3)
```

```python
random_forest_resampled = RandomForestClassifier(n_estimators=100)
```

```python
random_forest_resampled.fit(X_train, y_train)
```

```python
y_pred = random_forest_resampled.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
